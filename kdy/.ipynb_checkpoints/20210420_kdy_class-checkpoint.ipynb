```python

#-----------------------------------------------------------------------------
# csv 파일 로드
import csv



class CsvDATA: # csv 파일 로드 
    def __init__(self,fname):
        self.lists=[]
        file=open(fname,encoding='cp949') # 변수명=open(파일 경로,encoding 작업)
        getline = csv.reader(file)
        line_count=0
        for line in getline:
            if line_count>5:
                self.lists.append(line)
            line_count+=1
        file.close()
        
    def Counts_Genre(self,list_input,genre_list):
        count_list=[]
        for genre in genre_list:
            count_list.append(0)
        
        for i in list_input:
            for idx,name_input in enumerate(genre_list):
                if name_input in i[11]:
                    count_list[idx]+=1
        return count_list
    
    def GetYearList(self,years):
        Data_list_years=[]
        for year in years:
            Data_list_years.append([])

        datas=self.lists.copy()
        for data in enumerate(datas):
            for idx,year in enumerate(years):
                if year in data[2]:
                    Data_list_years[idx].append(data)
        return Data_list_years


```

```python
dataA3 = CsvDATA('Data_List_A3.csv')
```
