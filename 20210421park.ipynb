```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc#한글
import csv

list1=['액션','드라마','SF','어린이','판타지','스릴러','코미디','모험','전쟁','뮤지컬']
list2=[]
list3=[]
a=0
for k in range(2011,2021):#k 변수 2011~2021년도 설정    
    
    for j in range(10):#list1설정  
        f=open('Data_List_A3.csv',encoding='cp949')
        data = csv.reader(f)
        header=next(data)
        for i in data:
            if str(k) in i[2]:
                if list1[j] in i[11]:
                    a+=1    
        list2.append(a)
        f.close()
        a=0        
list3 = list_chunk(list2,10)
print(list3)
rc('font', family='HCR Dotum')#한글

#category_names = ['액션','드라마','SF','어린이','판타지','스릴러','재난','코미디']
results = {
    '2011': list3[0],
    '2012': list3[1],
    '2013': list3[2],
    '2014': list3[3],
    '2015': list3[4],
    '2016': list3[5],
    '2017': list3[6],
    '2018': list3[7],
    '2019': list3[8],
    '2020': list3[9]
}

def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(10.75, 8))#바탕크기조절(가로,세로)
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(list1, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.6,
                        label=colname, color=color)        
        xcenters = starts + widths / 2#1
        r, g, b, _ = color
        text_color = 'black' #2        
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(list1), bbox_to_anchor=(0, 1.1),
              loc='lower left', fontsize='small')
    ax.set_ylabel('연도')
    ax.set_title('박스오피스 장르')
    return fig, ax
survey(results, list1)
plt.show()
```
